```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2016_DATA

```python
# Creating file path and read
csv_path = "resources/president_2016.csv"
data_2016 = pd.read_csv(csv_path)
data_2016.head()
```

```python
data_2016 = data_2016[['cycle', 'branch', 'state', 'pollster', 'grade', 'population', 'adjpoll_clinton', 'adjpoll_trump' ]]
data_2016.head()
```

```python
data_2016.rename(columns= {'cycle': 'Cycle', 'branch': 'Branch', 'state': 'State', 'pollster': 'Pollster', 'grade': 'Grade', 'population': 'Population', 'adjpoll_clinton': 'Clinton_adj_poll', 'adjpoll_trump': 'Trump_adj_poll'}, inplace= True )
data_2016.head()
```

```python
# load to csv
data_2016.to_csv('data_2016.csv')
```

## 2020_DATA

```python
# Creating file path and read
csv_path = "resources/president_2020.csv"
data_2020 = pd.read_csv(csv_path)
data_2020.head()
```

```python
data_2020 = data_2020[['cycle', 'state', 'pollster', 'fte_grade', 'population', 'office_type', 'answer', 'candidate_name', 'pct']]
data_2020.head()
```

```python
data_2020.rename(columns={'cycle': 'Cycle', 'state': 'State', 'pollster': 'Pollster', 'fte_grade': 'FTE Grade', 'population': 'Population', 'office_type': 'Office Type', 'answer': 'Answer', 'candidate_name': 'Candidate Name', 'pct': 'Pct'}, inplace=True)
data_2020.head()
```
